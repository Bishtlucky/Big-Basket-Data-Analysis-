```python
# Step 5 : FIND OUT TOP 5 & LEAST 5 SELLING PRODUCT
# using group by
product_sale = df['product'].value_counts()
top_sales = product_sale.nlargest(5)
least_sale = product_sale.nsmallest(5)
```

```python
# Printing both outcomes
print("Top 5 selling Product")
print(top_sales)
print("\nLeast 5 selling Product")
print(least_sale)
```

```python
# Concatinate TOP and LEAST Product for Visualization
top_least_products =pd.concat([top_sales, least_sale])
# Showing a bar plot
plt.figure(figsize=(12,6))
sns.barplot(x=top_least_products.index, y=top_least_products.values)
plt.xticks(rotation=45 , ha="right")
plt.xlabel("product name")
plt.ylabel("solding quantity")
plt.title("TOP and LEAST 5 Products by sales count")
plt.tight_layout()
plt.show()
```

```python
# STEP 6 : Measuring discount on a certian item
df['Discount percentage'] = ((df['market_price'] - df['sale_price']) / df['market_price']) * 100

# Sort products by discount percentage (descending for top, ascending for least)
df_sorted = df.sort_values(by='Discount percentage', ascending=False)

# Get top 5 products with the highest discount
top_5 = df_sorted.head(5)

# Get least 5 products with the lowest discount
least_5 = df_sorted.tail(5)
```

```python
# Display the results
print("Top 5 Products with the Highest Discount:")
print(top_5[['product', 'market_price', 'sale_price', 'Discount percentage']])

print("\nLeast 5 Products with the Lowest Discount:")
print(least_5[['product','market_price', 'sale_price', 'Discount percentage']])
```

```python
# Step 8 : finding outliers
# creating a boxplot for sale price, market price, rating
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
sns.boxplot(y=df['sale_price'])
plt.title('Box plot for sale price')

plt.subplot(1,3,2)
sns.boxplot(y=df['market_price'])
plt.title('Box plot for market price')

plt.subplot(1,3,3)
sns.boxplot(y=df['rating'])
plt.title('Box plot for rating')

plt.tight_layout()
plt.show()
```

```python
# calculating thw IQR for the "sale_price"
Q1_sale = df['sale_price'].quantile(0.25)
Q3_sale = df['sale_price'].quantile(0.75)
IQR_sale = Q3_sale - Q1_sale

lower_bound_sale = Q1_sale - 1.5 * IQR_sale
upper_bound_sale = Q3_sale + 1.5 * IQR_sale

#replacing the ouliers with the mean
df['sale_price'] = np.where((df['sale_price'] < lower_bound_sale) | (df['sale_price'] > upper_bound_sale), df['sale_price'].mean(), df['sale_price'])
```

```python
Q1_sale = df['market_price'].quantile(0.25)
Q3_sale = df['market_price'].quantile(0.75)
IQR_sale = Q3_sale - Q1_sale

lower_bound_sale = Q1_sale - 1.5 * IQR_sale
upper_bound_sale = Q3_sale + 1.5 * IQR_sale

df['market_price'] = np.where((df['market_price'] < lower_bound_sale) | (df['market_price'] > upper_bound_sale), df['market_price'].mean(), df['market_price'])
```

```python
Q1_sale = df['rating'].quantile(0.25)
Q3_sale = df['rating'].quantile(0.75)
IQR_sale = Q3_sale - Q1_sale

lower_bound_sale = Q1_sale - 1.5 * IQR_sale
upper_bound_sale = Q3_sale + 1.5 * IQR_sale

df['rating'] = np.where((df['rating'] < lower_bound_sale) | (df['rating'] > upper_bound_sale), df['rating'].mean(), df['rating'])
```

```python
plt.figure(figsize =(15,5))

plt.subplot(1,3,1)
sns.boxplot(y = df['sale_price'])
plt.title('box plot of sale price')

plt.subplot(1,3,2)
sns.boxplot(y = df['market_price'])
plt.title('box plot of market price')

plt.subplot(1,3,3)
sns.boxplot(y = df['rating'])
plt.title('box plot of rating')

plt.tight_layout()
plt.show()
```
